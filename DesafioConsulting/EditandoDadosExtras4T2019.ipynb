```
j = 0
for dado in data['Despesa Hóspedes']:
  dado = dado.replace('R$', '')
  dado = dado.replace(' ', '')
  dado = dado.replace(',','.')
  data['Despesa Hóspedes'][j] = float(dado)
  j += 1

data['Despesa Hóspedes']
```
