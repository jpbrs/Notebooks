```
import pandas as pd
```

```
data = pd.read_csv('Estorno1T2020.csv')
```

```
data.head()
```

```
data = data.drop(columns = ['Unnamed: 0'], axis=1)
```

```
j = 0

for i in range(len(data[' Hotel - Valor a ser estornado'])):
  data[' Hotel - Valor a ser estornado'][j] = data[' Hotel - Valor a ser estornado'][j].split("-")
  j += 1

data.head()
```

```
data['Hotel'] = data[' Hotel - Valor a ser estornado']
data['Valor'] = data[' Hotel - Valor a ser estornado']
```

```
for i in range(len(data['Hotel'])):
  data['Hotel'][i] = data['Hotel'][i][0]
  data['Valor'][i] = data['Valor'][i][1]

data.head()
```

```
data = data.drop(columns = [' Hotel - Valor a ser estornado'], axis=1)
```

```
data.head()
```

```
for i in range(len(data['Valor'])):
  data['Valor'][i] = float(data['Valor'][i].replace(",","."))

data.head()
```
