```
import pandas as pd
```

```
data_reserva = pd.read_csv('NumeroReservaLocal.csv')
```

```
data_reserva.head()
```

```
data_reserva = data_reserva.drop(['Unnamed: 0'], axis=1)
```

```
data_reserva.Cidade.unique
```

```
rio_reservas = data_reserva.query('Cidade == "Rio de Janeiro"')
```

```
rio_reservas.head()
```

```
rio_reservas.describe()
```

```
sp_reservas = data_reserva.query('Cidade == "São Paulo"')
```

```
sp_reservas.head()
```

```
curitiba_reservas = data_reserva.query('Cidade == "Curitiba"')
```

```
curitiba_reservas.head()
```

```
bra_reservas = data_reserva.query('Cidade == "Brasília"')
bra_reservas.head()
```

```
data_custo = pd.read_csv('Extras4T2019.csv')
```

```
data_custo = data_custo.drop(columns='Unnamed: 0', axis = 1)
```

```
data_custo.head()
```

```
total_rj = 0

for reserva in rio_reservas['Número da reserva']:
  for i in range(len(data_custo['Número da reserva'])):
    if reserva == data_custo['Número da reserva'][i]:
      total_rj += data_custo['Despesa Hóspedes'][i]

total_rj
```

```
total_sp = 0

for reserva in sp_reservas['Número da reserva']:
  for i in range(len(data_custo['Número da reserva'])):
    if reserva == data_custo['Número da reserva'][i]:
      total_sp += data_custo['Despesa Hóspedes'][i]

total_sp
```

```
total_curitiba = 0

for reserva in curitiba_reservas['Número da reserva']:
  for i in range(len(data_custo['Número da reserva'])):
    if reserva == data_custo['Número da reserva'][i]:
      total_curitiba += data_custo['Despesa Hóspedes'][i]

total_curitiba
```

```
total_brasilia = 0

for reserva in bra_reservas['Número da reserva']:
  for i in range(len(data_custo['Número da reserva'])):
    if reserva == data_custo['Número da reserva'][i]:
      total_brasilia += data_custo['Despesa Hóspedes'][i]

total_brasilia
```

```
total_brasilia + total_curitiba + total_rj + total_sp
```
