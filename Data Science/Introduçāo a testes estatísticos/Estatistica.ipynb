```
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```
uri = "tmdb_5000_movies.csv"

tmdb = pd.read_csv(uri)
```

```
tmdb.head()
```

```
tmdb.describe()
```

```
sb.distplot(tmdb.vote_average)
```

```
isnull = tmdb.vote_average.isnull()
isnull.unique()
```
