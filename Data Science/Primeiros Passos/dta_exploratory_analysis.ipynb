```
  import pandas as pd
import seaborn as sb
```

```
uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/aula0/aula0/ml-latest-small/ratings.csv"
dados = pd.read_csv(uri)
dados.head()
```

```
mapa = {"userId":"id_usuario", "movieId":"id_filme", "rating":"nota", "timestamp":"tempo"}

dados = dados.rename(columns = mapa)
dados.head()
```

```
dados.nota.plot(kind='hist')
```

```
dados.nota.mean()
```

```
dados.nota.value_counts()
```

```
dados.nota.describe()
```

```
sb.boxplot(dados.nota)
```

```
uri2 = 'https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/aula0/aula0/ml-latest-small/movies.csv'
filmes = pd.read_csv(uri2)
filmes.head()
```

```
mapa = {"movieId": "id_filme", "title":"titulo", "genres":"generos"}
filmes = filmes.rename(columns = mapa)
filmes.head()
```

```
dados.query("id_filme==1").head()
```

```
dados.query("id_filme==1").nota.mean()
```

```
notas_filmes = dados.groupby("id_filme").mean() #agrupa por id_filme e tira a media das notas de tudo
notas_filmes.head()
```

```
notas_filmes.nota.plot(kind="hist")
```

```
sb.distplot(notas_filmes.nota, bins = 10)
```
