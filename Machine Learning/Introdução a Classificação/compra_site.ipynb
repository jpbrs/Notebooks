```
import sklearn
import numpy as np
import pandas as pd
import seaborn as sb
import scipy
```

```
data = pd.read_csv('https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv')
```

```
data.head()
```

```
mapa = { "home" : "Página Principal",
        "how_it_works" : "Como Funciona",
        "contact" : "Contato",
        "bought" : "Comprou"
}

data = data.rename(columns = mapa)
```

```
X=data[["Página Principal",
        "Como Funciona",
        "Contato"]]
Y=data[["Comprou"]]
```

```
data.shape
```

```
from sklearn.svm import LinearSVC
from sklearn.metrics import *
from sklearn.model_selection import *
```

```
RANDOM_SEED = 20 #serve para tirarmos a aleatoriadade do train test split
```

```
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, random_state = RANDOM_SEED ,stratify = Y, test_size = 0.25)#stratify serve para o treino e o teste ficarem proporcionais
```

```
model = LinearSVC()
model.fit(X_treino, Y_treino)
```

```
predict = model.predict(X_teste)
accuracy_score(Y_teste, predict)
```
